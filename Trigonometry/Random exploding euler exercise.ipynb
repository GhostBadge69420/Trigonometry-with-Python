```python
nvects = 200
ms = np.random.rand(nvects)
ks = np.random.rand(nvects)*2*np.pi

for i in range(0,nvects):
    
    r = np.random.rand()

    if r<.4:
        clr = [1,.2,.7]
    elif r>.4 and r<.8:
        clr = [.7,.2,1]
    else:
        clr = [0,1,0]

    plt.polar([0,ks[i]],[0,ms[i]],color=clr)

plt.axis('off')
plt.show()
```

```python
nvecs= 200

ms= np.random.rand(nvecs)

ks= np.random.rand(nvecs)*2*np.pi

x = ms*np.cos(ks)

y = ms*np.sin(ks)

for i in range(nvecs):

    if i<= 80:

        plt.plot([0,x[i]],[0,y[i]],color = [1,.2,.7])

    elif i>80 and i<=160:

        plt.plot([0,x[i]],[0,y[i]],color = [.7,.2,1])

    else:

        plt.plot([0,x[i]],[0,y[i]],color =  [0,1,0])

plt.axis("square")

plt.axis("off")

plt.show()
```
