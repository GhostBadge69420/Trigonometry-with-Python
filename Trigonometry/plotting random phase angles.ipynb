# Exercise

Plot unit vectors with random phase angles.

What random distributions are most sensible?

```python
angs = np.random.rand(500)*2*np.pi

for i in angs:
    plt.plot([0,np.cos(i)],[0,np.sin(i)])

plt.axis('square')
plt.axis([-1,1,-1,1])
plt.axis('off')
plt.show()
```
