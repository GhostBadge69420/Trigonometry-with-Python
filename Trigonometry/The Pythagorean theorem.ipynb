```python
# Generate a random complex number between (-10:10, -10:10j)
 
value = np.complex(((np.random.rand(1)*2)-1)*10, ((np.random.rand(1)*2)-1)*10)
 
##test## value = np.complex(3,2)
 
# absolute values for trig 
 
a = abs(np.real(value))
b = abs(np.imag(value))
 
# Pythagorean Theorem
 
c = np.sqrt(a**2 + b**2)
 
# Plot lines and points
 
# c line (red)
 
plt.plot(np.real(value), np.imag(value), 'ro', label= f'val = {value:3.2f}')
plt.plot((0, np.real(value)), (0, np.imag(value)), 'r-', label= f'c = {c:0.3f}')
 
# b line (blue, x axis)
 
plt.plot(np.real(value), 0, 'bo', label = f'{np.real(value):3.2f}, 0j')
plt.plot((np.real(value), np.real(value)), (0, np.imag(value)), 'b--', label = f'b = {b:0.3f}', alpha = 0.6)
 
# a line (green, y axis)
 
plt.plot(0, np.imag(value), 'go', label = f'0, {np.imag(value):3.2f}j')
plt.plot((0, np.real(value)), (np.imag(value), np.imag(value)), 'g--', label = f'a = {a:0.3f}', alpha = 0.6)
 
# dot for 0,0
 
plt.plot(0,0, 'ko', label = '0, 0')
 
# Plot axis lines, dotted
 
plt.plot([0,0], [-10,10], 'k--', alpha = 0.5)
plt.plot([-10,10], [0,0], 'k--', alpha = 0.5)
 
# Decoration
 
plt.axis('square')
plt.axis([-10,10,-10,10])
 
plt.grid()
plt.xlabel('real')
plt.ylabel('imag')
 
plt.legend()
plt.gca().legend(bbox_to_anchor=(1.4, 1.05))
plt.savefig('output.pdf', format='pdf', dpi=200)
plt.show()
 
# Calculate angle 
 
blueblackred = np.arctan(b/a)
greenblackred = np.pi/2 - blueblackred
 
blueredblack = np.pi - (blueblackred + np.pi/2)
greenredblack = (np.pi/2) - blueredblack
 
# Angles
 
print(f'Angle blue, black, red is {blueblackred} radians and {np.rad2deg(blueblackred)} degrees')
print(f'Angle green, back red is {greenblackred} radians and {np.rad2deg(greenblackred)} degrees')
print('\n')
print(f'Angle blue, red, black is {blueredblack} radians and {np.rad2deg(blueredblack)} degrees')
print(f'Angle green, red, black is {greenredblack} radians and {np.rad2deg(greenredblack)} degrees')
```
