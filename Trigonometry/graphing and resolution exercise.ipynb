```python
def format_func(value, tick_number):
    # find number of multiples of pi/2
    N = int(np.round(2 * value / np.pi))
    if N == 0:
        return "0"
    elif N == -2:
        return r"$-\pi$"
    elif N == -1:
      return r"$\frac{-\pi}{2}$"
    elif N == 1:
        return r"$\frac{\pi}{2}$"
    elif N == 2:
        return r"$\pi$"
    elif N % 2 > 0:
        return r"$\frac{{{0}\pi}}{{2}}$".format(N)
    else:
        return r"${0}\pi$".format(N // 2)
 
angles = np.linspace(-2*np.pi, 2*np.pi, 361)
 
y1 = np.sin(angles + np.cos(angles))
y2 = np.cos(angles + np.sin(angles))
 
trig = sns.lineplot(x=angles, y=y1)
plt.plot(angles, y2)
trig.xaxis.set_major_locator(plt.MultipleLocator(np.pi / 2))
trig.xaxis.set_minor_locator(plt.MultipleLocator(np.pi / 6))
trig.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
plt.legend(["$\\sin(\\theta + \\cos(\\theta))$",
           "$\\cos(\\theta + \\sin(\\theta))$"])
plt.show()
```
