# Exercise

Generate a normal distribution with:

mean = 14

standard deviation = 4.3

```python
desired_mean = 15
desired_std = 4.3

nums = np.random.randn(1000)
nums = nums - np.mean(nums)
nums = nums/np.std(nums)
nums = nums*desired_std + desired_mean

plt.subplot(121)
plt.plot(nums,'s')

plt.subplot(122)
plt.hist(nums)

plt.subplot(122)
plt.hist(nums)
plt.show()

print('This distribution has a mean of %s anbd a standard deviation of %s' %(np.mean(nums),np.std(nums)))
```
