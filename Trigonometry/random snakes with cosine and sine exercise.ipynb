```python
angs = np.linspace(0,20*np.pi,1000)

#cs = np.random.rand()
#ss = np.random.rand()

cs = 8
ss = 9

x = np.sin(cs*angs)
y = np.sin(ss*angs)

# x = np.append(x,x[0])
# y = np.append(y,y[0])

plt.plot(x,y,'k')
plt.axis('square')
plt.axis([-1.1,1.1,-1.1,1.1])
plt.title('r$_1$ = %s, r$_2$ = %s' %(np.round(cs,2),np.round(ss,2)))
plt.axis('off')
plt.show()
```
